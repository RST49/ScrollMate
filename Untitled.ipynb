```python
import cv2
import mediapipe as mp
```

```python
mpHands = mp.solutions.hands
hands = mpHands.Hands(
    static_image_mode=False,
    model_complexity=1,
    min_detection_confidence=0.7,
    min_tracking_confidence=0.7,
    max_num_hands=1
)
```

```python
def main():
    cap = cv2.VideoCapture(0)
    draw = mp.solutions.drawing_utils
    try:
        while cap.isOpened():
            ret , frame = cap.read()
            if not ret:
                break;
            frame = cv2.flip(frame,1)
            frameCol = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
            processed = hands.process(frameCol)
            land_list = []

            if processed.multi_hand_landmarks:
                hand_land = processed.multi_hand_landmarks[0]
                draw.draw_landmarks(frame,hand_land,mpHands.HAND_CONNECTIONS)

                for lm in hand_land.landmark:
                    land_list.append((lm.x,lm.y))
                
            
            cv2.imshow('Frame', frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break;

    finally:
        cap.release()
        cv2.destroyAllWindows()


main()
```
